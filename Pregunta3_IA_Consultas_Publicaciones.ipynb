$\large\textit{6. Integrando la Consulta con la Base de Datos}$

```python
import pandas as pd

# Importando el archivo CSV desde GitHub
url = 'https://raw.githubusercontent.com/EHN8829/DUPD_PARCIAL/main/scopus.csv'
```

```python
# Cargando el archivo CSV en un DataFrame de pandas
df = pd.read_csv(url)
```

```python
# Verificando las primeras filas del DataFrame
print(df.head())
```

```python
# Verificando las primeras filas del DataFrame como tabla
df.head()
```

```python
import torch
```

```python
# Definiendo la consulta
consulta = "artículos sobre redes neuronales y aprendizaje profundo en biomedicina"
```

```python
# Tokenizando la consulta
inputs = tokenizer(consulta, return_tensors="pt")
```

```python
# Generando la respuesta usando el modelo
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=150)
```

```python
# Decodificando la respuesta
respuesta = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Consulta:", consulta)
print("Respuesta:", respuesta)
```

```python
# Filtrando la base de datos basada en la consulta, por ejemplo, en el campo del título o resumen
df_filtrado = df[df['Title'].str.contains("deep learning", case=False, na=False)]
```

```python
# Verificando las primeras filas del DataFrame
print(df_filtrado[['Title', 'Abstract']].head())
```

```python
# Verificando las primeras filas del DataFrame en forma de tabla
df_filtrado[['Title', 'Abstract']].head()
```

$\large\textit{7. Respondiendo Consultas Dinámicas Basadas en el Modelo}$

```python
# Ejemplo de pregunta
pregunta = "¿Qué artículos sobre aprendizaje profundo en biomedicina usan redes neuronales convolucionales?"
```

```python
# Tokenizando y generando una respuesta
inputs = tokenizer(pregunta, return_tensors="pt")
outputs = model.generate(**inputs, max_length=150)
respuesta = tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print("Pregunta:", pregunta)
print("Respuesta del modelo:", respuesta)
```

```python
# Filtrando en la base de datos de acuerdo a la respuesta o la consulta
df_filtrado = df[df['Abstract'].str.contains("convolutional neural networks", case=False, na=False)]
```

```python
# Mostrando los artículos relevantes
print(df_filtrado[['Title', 'Abstract']].head())
```

```python
# Presentando los resultados relevantes en forma de tabla
df_filtrado.head()
```

$\large\textit{8. Asignando un Indicador Numérico (Puntaje) y Filtrando el Dataset}$

```python
# Asignando un puntaje basado en la longitud del título
df['puntaje'] = df['Title'].apply(lambda x: len(x) if isinstance(x, str) else 0)
```

```python
# Mostrando los primeros registros para su verificación
print(df[['Title', 'puntaje']].head())
```

```python
# Mostrando los primeros registros para verificación en forma de tabla
df[['Title', 'puntaje']].head()
```

```python
# Filtrando el dataset basado en el puntaje
# Por ejemplo, quedándonos con los registros cuyo puntaje sea menor que 50
df_filtrado = df[df['puntaje'] < 50]
```

```python
# Mostrando las primeras filas del DataFrame filtradas para su verificación
print(df_filtrado[['Title', 'puntaje']].head())
```

```python
# Mostrando las primeras filas del DataFrame filtradas para su verificación en forma de tabla
df_filtrado[['Title', 'puntaje']].head()
```
